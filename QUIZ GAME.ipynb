```python
import random

# Quiz questions in the format: (question, correct_answer, [answer_choices])
quiz_questions = [
    ("What is the capital of France?", "Paris", ["London", "Berlin", "Madrid", "Paris"]),
    ("Which planet is known as the 'Red Planet'?", "Mars", ["Venus", "Mars", "Jupiter", "Mercury"]),
    ("What is the chemical symbol for water?", "H2O", ["H2O", "CO2", "O2", "N2"]),
    ("Who painted the Mona Lisa?", "Leonardo da Vinci", ["Pablo Picasso", "Vincent van Gogh", "Leonardo da Vinci", "Michelangelo"]),
]

def load_question():
    return random.choice(quiz_questions)

def display_welcome_message():
    print("Welcome to the Quiz Game!")
    print("You will be asked multiple-choice questions. Choose the correct answer by typing the letter (A, B, C, D).")
    print("Let's begin!")

def prompt_user_answer(choices):
    while True:
        user_input = input("Enter your answer (A, B, C, D): ").strip().upper()
        if user_input in choices:
            return user_input
        else:
            print("Invalid choice. Please select a valid option.")

def evaluate_answer(user_answer, correct_answer):
    if user_answer == correct_answer:
        print("Correct!")
        return True
    else:
        print(f"Incorrect. The correct answer is: {correct_answer}")
        return False

def calculate_final_score(score, total_questions):
    return (score / total_questions) * 100

def display_final_results(score, total_questions):
    final_score = calculate_final_score(score, total_questions)
    print(f"\nYour Final Score: {final_score:.2f}%")
    if final_score == 100:
        print("Congratulations! You got a perfect score!")
    elif final_score >= 70:
        print("Well done! You did a great job!")
    else:
        print("Keep practicing. You can do better next time!")

def play_quiz_game():
    display_welcome_message()
    total_questions = 0
    score = 0

    while True:
        question, correct_answer, answer_choices = load_question()
        print(f"\nQuestion {total_questions + 1}: {question}")
        for i, choice in enumerate(answer_choices):
            print(f"  {chr(65 + i)}. {choice}")

        user_answer = prompt_user_answer([chr(65 + i) for i in range(len(answer_choices))])
        total_questions += 1
        if evaluate_answer(user_answer, correct_answer):
            score += 1

        play_again = input("\nDo you want to play again? (yes/no): ").strip().lower()
        if play_again != "yes":
            break

    display_final_results(score, total_questions)

if __name__ == "__main__":
    play_quiz_game()
```
